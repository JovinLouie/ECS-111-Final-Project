```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Iaine/.vscode/Python Projects/riot-user-scraper/code/valorant_players_stats_final - valorant_players.csv")
df
```

```python
df = df.dropna(subset=['hs_percent', 'body_percent', 'leg_percent',
       's_damage_per_round', 's_kd_ratio', 's_hs_percent', 's_win_percent',
       's_wins', 's_kast_percent', 's_dd_round', 's_kills', 's_deaths',
       's_assists', 's_acs', 's_kad_ratio', 's_kills_per_round',
       's_first_bloods', 's_flawless_rounds', 's_aces'])
```

```python
columns_with_percent = ['hs_percent', 'body_percent', 'leg_percent', 's_hs_percent', 's_win_percent', 's_kast_percent']
for col in columns_with_percent:
    df[col] = df[col].str.replace('%', '', regex=False)
```

```python
df = df.applymap(lambda x: x.rstrip('.') if isinstance(x, str) else x)

df = df.applymap(lambda x: x.replace(',', '') if isinstance(x, str) else x)
```

```python
columns_to_float64 = list(df.columns)
for col_to_remove in ['puuid', 'user', 'tag']:
	columns_to_float64.remove(col_to_remove)
	
for column in columns_to_float64:
	df[column] = df[column].astype(np.float64)

df.dtypes
```

```python
df
```

```python
df['deaths_per_game'] = df['s_deaths'] / (df['s_wins'] + ((df['s_wins'] / df['s_win_percent']) * (100 - df['s_win_percent'])))

df['kills_per_game'] = df['s_kills'] / (df['s_wins'] + ((df['s_wins'] / df['s_win_percent']) * (100 - df['s_win_percent'])))

df['assists_per_game'] = df['s_assists'] / (df['s_wins'] + ((df['s_wins'] / df['s_win_percent']) * (100 - df['s_win_percent'])))

df['first_bloods_per_game'] = df['s_first_bloods'] / (df['s_wins'] + ((df['s_wins'] / df['s_win_percent']) * (100 - df['s_win_percent'])))

df['flawless_rounds_per_game'] = df['s_flawless_rounds'] / (df['s_wins'] + ((df['s_wins'] / df['s_win_percent']) * (100 - df['s_win_percent'])))

df['aces_per_game'] = df['s_aces'] / (df['s_wins'] + ((df['s_wins'] / df['s_win_percent']) * (100 - df['s_win_percent'])))

df = df.drop(columns=['s_deaths', 's_kills', 's_assists', 's_wins','s_first_bloods', 's_flawless_rounds', 's_aces'])

df
```

```python
# First calculate per-game stats
'''df['games_played'] = df['wins'] + df['losses']
df['kills_per_game'] = df['kills'] / df['games_played']
df['deaths_per_game'] = df['deaths'] / df['games_played']
df['assists_per_game'] = df['assists'] / df['games_played']
df['first_bloods_per_game'] = df['first_bloods'] / df['games_played']
df['flawless_per_game'] = df['flawless_rounds'] / df['games_played']
df['aces_per_game'] = df['aces'] / df['games_played']'''

df['smurf_label'] = 'normal player'

for idx, row in df.iterrows():
    smurf_counter = 0
    sus_counter = 0

    if row['s_damage_per_round'] > 160:
        smurf_counter += 1
    elif row['s_damage_per_round'] > 150:
        sus_counter += 1

    if row['s_kd_ratio'] > 1.3:
        smurf_counter += 1
    elif row['s_kd_ratio'] > 1.1:
        sus_counter += 1

    if row['s_win_percent'] > 60:
        smurf_counter += 1
    elif row['s_win_percent'] > 50:
        sus_counter += 1

    if row['hs_percent'] > 25:
        smurf_counter += 1
    elif row['hs_percent'] > 22:
        sus_counter += 1

    if row['s_acs'] > 250:
        smurf_counter += 1
    elif row['s_acs'] > 220:
        sus_counter += 1

    if row['s_dd_round'] > 10:
        smurf_counter += 1
    elif row['s_dd_round'] > 5:
        sus_counter += 1

    if row['s_kills_per_round'] > 0.9:
        smurf_counter += 1
    elif row['s_kills_per_round'] > 0.7:
        sus_counter += 1

    if row['leg_percent'] < 8:
        smurf_counter += 1
    elif row['leg_percent'] < 9:
        sus_counter += 1

    if row['kills_per_game'] > 20:
        smurf_counter += 1
    elif row['kills_per_game'] > 17:
        sus_counter += 1

    if row['deaths_per_game'] < 12:
        smurf_counter += 1
    elif row['deaths_per_game'] < 15:
        sus_counter += 1

    if row['assists_per_game'] > 5:
        smurf_counter += 1
    elif row['assists_per_game'] > 4:
        sus_counter += 1

    if row['first_bloods_per_game'] > 3:
        smurf_counter += 1
    elif row['first_bloods_per_game'] > 2:
        sus_counter += 1

    if row['flawless_rounds_per_game'] > 1.2:
        smurf_counter += 1
    elif row['flawless_rounds_per_game'] > 0.8:
        sus_counter += 1

    if row['aces_per_game'] > 0.05:
        smurf_counter += 1
    elif row['aces_per_game'] > 0.02:
        sus_counter += 1

    if smurf_counter > 4:
        df.at[idx, 'smurf_label'] = 'most likely smurf'
    elif sus_counter > 4:
        df.at[idx, 'smurf_label'] = 'suspicious'
```
